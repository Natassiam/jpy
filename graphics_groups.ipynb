**Grupos gerados na etapa de agrupamento**

Importação das bibliotecas utilizadas para gerar os gráficos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nomes = ['baixo', 'médio', 'alto']
```

```python
quant = [147, 343, 126]
```

```python
cores = ['#544d64', '#472a87',  '#3a0ba6' ]
```

```python
plt.title('Grupo 4')
plt.bar(nomes, quant, color=cores)
for i in range(len(nomes)):
    plt.text(nomes[i], quant[i]+0.2, quant[i], ha='center',  fontsize=9)
    plt.savefig('g4.png', format='png')
```
