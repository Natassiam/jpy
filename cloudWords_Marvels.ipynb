```python
all_urlslug = " ".join(s for s in urlslug)
```

```python
stopwords = set(STOPWORDS)
stopwords.update(["_", "//", "%", "22", "Reborn", "Doctor", "os", "Earth"])
```

```python
wordcloud = WordCloud(stopwords=stopwords,
                      background_color="black",
                      width=1600, height=800).generate(all_urlslug)
 
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()
 
plt.imshow(wordcloud);
wordcloud.to_file("marvels_name_wordcloud.png")
```
